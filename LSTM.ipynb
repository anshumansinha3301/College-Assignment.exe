# ğŸ§  Long Short-Term Memory (LSTM)


## ğŸ“˜ Introduction

A **Long Short-Term Memory (LSTM)** network is a special kind of **Recurrent Neural Network (RNN)** capable of learning **long-term dependencies** in sequential data.  
It was introduced by *Hochreiter and Schmidhuber (1997)* to solve the **vanishing gradient problem** that limits vanilla RNNs.

LSTMs can **remember information over long sequences** using an internal memory cell and a system of **gates** that control the flow of information.

---

## âš™ï¸ Architecture Overview

An LSTM unit has four main components:  
1. **Forget Gate** â€” decides what to remove from memory.  
2. **Input Gate** â€” decides what new information to add.  
3. **Cell State** â€” stores long-term information.  
4. **Output Gate** â€” decides what to output from memory.

---

## ğŸ” Flow of Information

1. The **Forget Gate** clears irrelevant information.  
2. The **Input Gate** updates the memory cell with new input.  
3. The **Cell State** carries long-term information forward.  
4. The **Output Gate** produces the next hidden state based on memory.

This gating mechanism allows **controlled information flow**, helping the LSTM remember useful data and forget noise.

---

## ğŸ§© Key Characteristics

- **Memory cell** separates short- and long-term memory.  
- **Three gates (f, i, o)** dynamically regulate memory updates.  
- **Resistant to vanishing gradients**, enabling long-range learning.  
- **Shared weights** across time steps (like RNNs).  

---

## ğŸ“Š Types of LSTM Architectures

| Type | Description | Example Use |
|:------|:-------------|:-------------|
| **Vanilla LSTM** | Single-layer, unidirectional | Text sentiment classification |
| **Stacked LSTM** | Multiple LSTM layers stacked | Speech recognition |
| **Bidirectional LSTM** | Processes input from both directions | Language modeling |
| **Encoderâ€“Decoder LSTM** | Used in seq2seq models | Machine translation |

---

## ğŸ§® LSTM Variants

| Variant | Description | Advantage |
|:----------|:-------------|:-----------|
| **Vanilla LSTM** | Standard form with 3 gates | Long-term memory retention |
| **Peephole LSTM** | Gates access cell state directly | Finer timing control |

| **Stacked LSTM** | Multiple layers of LSTM cells | Deep temporal understanding |
| **ConvLSTM** | Convolutional operations inside LSTM | Video and spatiotemporal data |

---

## ğŸ§  Advantages

âœ… Solves vanishing gradient problem  
âœ… Learns both long- and short-term dependencies  
âœ… Flexible and robust for variable-length sequences  
âœ… Works well in noisy or incomplete sequences  
âœ… Strong generalization for temporal data  

---

## âš ï¸ Limitations

âŒ Computationally expensive (4Ã— parameters vs RNN)  
âŒ Slower to train than GRU or RNN  
âŒ Difficult to parallelize across timesteps  
âŒ High memory usage for large sequences  

---

## ğŸ§© Improvements & Enhancements

1. **Bidirectional LSTM** â€“ improves context comprehension.  
2. **Attention Mechanism** â€“ dynamically weights important timesteps.  
3. **ConvLSTM** â€“ integrates spatial (CNN) and temporal (LSTM) features.  
4. **Residual LSTM** â€“ stabilizes deep stacked LSTMs.  
5. **Hybrid Models** â€“ LSTM + CNN or LSTM + Transformer hybrids.

---

## ğŸ”¬ Training: Backpropagation Through Time (BPTT)

LSTM training uses **BPTT**, unrolling across time and updating weights for all gates.

**Advantages over RNN:**
- Gradient flow is preserved via cell state \( C_t \).  
- Forget gate helps retain useful memory and discard noise.  

**Regularization techniques:**  
- Dropout between LSTM layers  
- Gradient clipping  
- Early stopping  

---

## ğŸ§¾ Activation Functions Used

| Function | Equation | Use |
|:----------|:-----------|:----|
| **Sigmoid (Ïƒ)** | \( \frac{1}{1 + e^{-x}} \) | Used for gates (output between 0 and 1) |
| **Tanh** | \( \frac{e^x - e^{-x}}{e^x + e^{-x}} \) | Used for cell state and hidden updates |
| **ReLU** | \( \max(0, x) \) | Occasionally used for faster convergence |

# Comparative Implementation of Vanilla LSTM and Elman Networks

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, SimpleRNN, Dense, Embedding
from tensorflow.keras.utils import plot_model
import numpy as np

vocab_size = 10000
embedding_dim = 64
max_length = 100
hidden_units = 32
batch_size = 10

print("--- Building Vanilla LSTM Model ---")
model_lstm = Sequential(name="Vanilla_LSTM_Model")
model_lstm.add(Embedding(input_dim=vocab_size,
                         output_dim=embedding_dim,
                         input_length=max_length))
model_lstm.add(LSTM(units=hidden_units))
model_lstm.add(Dense(1, activation='sigmoid'))

model_lstm.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model_lstm.build(input_shape=(None, max_length))

print(f"This model uses the 'LSTM' layer.\n")
model_lstm.summary()

plot_model(model_lstm, to_file='lstm_model_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved LSTM diagram to 'lstm_model_diagram.png'")


print("\n" + "="*40 + "\n")

print("--- Building Elman (SimpleRNN) Model ---")
model_rnn = Sequential(name="Elman_SimpleRNN_Model")
model_rnn.add(Embedding(input_dim=vocab_size,
                        output_dim=embedding_dim,
                        input_length=max_length))
model_rnn.add(SimpleRNN(units=hidden_units))
model_rnn.add(Dense(1, activation='sigmoid'))

model_rnn.compile(
    loss='binary_crossentropy',
    optimizer='adam',
    metrics=['accuracy']
)

model_rnn.build(input_shape=(None, max_length))

print(f"This model uses the 'SimpleRNN' layer.\n")
model_rnn.summary()

plot_model(model_rnn, to_file='rnn_model_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved RNN diagram to 'rnn_model_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Creating Dummy Data ---")
dummy_x = np.random.randint(0, vocab_size, size=(batch_size, max_length))
dummy_y = np.random.randint(0, 2, size=(batch_size, 1))
print(f"Shape of input data: {dummy_x.shape}")
print(f"Shape of output data: {dummy_y.shape}")

print("\n" + "="*40 + "\n")

print("--- Training Vanilla LSTM Model ---")
model_lstm.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size)

print("\n" + "="*40 + "\n")

print("--- Training Elman (SimpleRNN) Model ---")
model_rnn.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size)

print("\nTraining complete.")
```

# Encoderâ€“Decoder LSTM (Seq2Seq)

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import LSTM, SimpleRNN, Dense, Embedding, Bidirectional, Input, TimeDistributed
from tensorflow.keras.utils import plot_model
import numpy as np

vocab_size = 10000
embedding_dim = 64
max_length = 100
hidden_units = 32
batch_size = 10

print("--- Building Elman (SimpleRNN) Model ---")
model_rnn = Sequential(name="Elman_SimpleRNN_Model")
model_rnn.add(Embedding(input_dim=vocab_size,
                        output_dim=embedding_dim,
                        input_length=max_length))
model_rnn.add(SimpleRNN(units=hidden_units))
model_rnn.add(Dense(1, activation='sigmoid'))

model_rnn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_rnn.build(input_shape=(None, max_length))

print(f"This model uses the 'SimpleRNN' layer.\n")
model_rnn.summary()

plot_model(model_rnn, to_file='rnn_model_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved RNN diagram to 'rnn_model_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Building Vanilla (Single-Layer) LSTM Model ---")
model_lstm_single = Sequential(name="Vanilla_LSTM_Model")
model_lstm_single.add(Embedding(input_dim=vocab_size,
                         output_dim=embedding_dim,
                         input_length=max_length))
model_lstm_single.add(LSTM(units=hidden_units))
model_lstm_single.add(Dense(1, activation='sigmoid'))

model_lstm_single.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_lstm_single.build(input_shape=(None, max_length))

print(f"This model uses one 'LSTM' layer.\n")
model_lstm_single.summary()

plot_model(model_lstm_single, to_file='lstm_single_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved Single-Layer LSTM diagram to 'lstm_single_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Building Stacked (Deep) LSTM Model ---")
model_lstm_stacked = Sequential(name="Stacked_LSTM_Model")

model_lstm_stacked.add(Embedding(input_dim=vocab_size,
                                 output_dim=embedding_dim,
                                 input_length=max_length))

model_lstm_stacked.add(LSTM(units=hidden_units, return_sequences=True))
model_lstm_stacked.add(LSTM(units=hidden_units))
model_lstm_stacked.add(Dense(1, activation='sigmoid'))

model_lstm_stacked.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_lstm_stacked.build(input_shape=(None, max_length))

print(f"This model uses two stacked 'LSTM' layers.\n")
model_lstm_stacked.summary()

plot_model(model_lstm_stacked, to_file='lstm_stacked_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved Stacked LSTM diagram to 'lstm_stacked_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Building Bidirectional LSTM (BiLSTM) Model ---")
model_bilstm = Sequential(name="Bidirectional_LSTM_Model")

model_bilstm.add(Embedding(input_dim=vocab_size,
                           output_dim=embedding_dim,
                           input_length=max_length))

model_bilstm.add(Bidirectional(LSTM(units=hidden_units)))
model_bilstm.add(Dense(1, activation='sigmoid'))

model_bilstm.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_bilstm.build(input_shape=(None, max_length))

print(f"This model uses one 'Bidirectional(LSTM)' layer.\n")
model_bilstm.summary()

plot_model(model_bilstm, to_file='bilstm_model_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved BiLSTM diagram to 'bilstm_model_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Building Encoder-Decoder LSTM (Seq2Seq) Model ---")

encoder_inputs = Input(shape=(max_length,), name="Encoder_Input")
enc_emb = Embedding(input_dim=vocab_size,
                    output_dim=embedding_dim,
                    input_length=max_length,
                    name="Encoder_Embedding")(encoder_inputs)

encoder_lstm = LSTM(hidden_units, return_state=True, name="Encoder_LSTM")
encoder_outputs, state_h, state_c = encoder_lstm(enc_emb)
encoder_states = [state_h, state_c]

decoder_inputs = Input(shape=(max_length,), name="Decoder_Input")

dec_emb_layer = Embedding(input_dim=vocab_size,
                          output_dim=embedding_dim,
                          input_length=max_length,
                          name="Decoder_Embedding")
dec_emb = dec_emb_layer(decoder_inputs)

decoder_lstm = LSTM(hidden_units, return_sequences=True, return_state=False, name="Decoder_LSTM")
decoder_outputs = decoder_lstm(dec_emb, initial_state=encoder_states)

decoder_dense = TimeDistributed(Dense(1, activation='sigmoid'), name="Time_Distributed_Dense")
decoder_outputs = decoder_dense(decoder_outputs)

model_seq2seq = Model([encoder_inputs, decoder_inputs], decoder_outputs, name="Encoder_Decoder_Model")

model_seq2seq.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print(f"This model uses the Functional API for Seq2Seq.\n")
model_seq2seq.summary()

plot_model(model_seq2seq, to_file='seq2seq_model_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved Seq2Seq diagram to 'seq2seq_model_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Creating Dummy Data for Models 1-4 ---")
dummy_x = np.random.randint(0, vocab_size, size=(batch_size, max_length))
dummy_y = np.random.randint(0, 2, size=(batch_size, 1))
print(f"Shape of standard input data (dummy_x): {dummy_x.shape}")
print(f"Shape of standard output data (dummy_y): {dummy_y.shape}")

print("\n" + "="*40 + "\n")

print("--- Creating Dummy Data for Seq2Seq Model 5 ---")
encoder_input_data = np.random.randint(0, vocab_size, size=(batch_size, max_length))
decoder_input_data = np.random.randint(0, vocab_size, size=(batch_size, max_length))
decoder_target_data = np.random.randint(0, 2, size=(batch_size, max_length, 1))
print(f"Shape of Seq2Seq encoder input: {encoder_input_data.shape}")
print(f"Shape of Seq2Seq decoder input: {decoder_input_data.shape}")
print(f"Shape of Seq2Seq decoder output: {decoder_target_data.shape}")

print("\n" + "="*40 + "\n")

print("--- Training Elman (SimpleRNN) Model ---")
model_rnn.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size, verbose=1)
print("SimpleRNN training complete.")

print("\n" + "="*40 + "\n")

print("--- Training Vanilla (Single-Layer) LSTM Model ---")
model_lstm_single.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size, verbose=1)
print("Single-Layer LSTM training complete.")

print("\n" + "="*40 + "\n")

print("--- Training Stacked (Deep) LSTM Model ---")
model_lstm_stacked.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size, verbose=1)
print("Stacked LSTM training complete.")

print("\n" + "="*40 + "\n")

print("--- Training Bidirectional LSTM (BiLSTM) Model ---")
model_bilstm.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size, verbose=1)
print("BiLSTM training complete.")

print("\n" + "="*40 + "\n")

print("--- Training Encoder-Decoder LSTM (Seq2Seq) Model ---")
model_seq2seq.fit([encoder_input_data, decoder_input_data], decoder_target_data,
                  epochs=3,
                  batch_size=batch_size,
                  verbose=1)
print("Seq2Seq training complete.")
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import LSTM, SimpleRNN, Dense, Embedding, Bidirectional, Input, TimeDistributed, Attention, GlobalAveragePooling1D, Conv1D, MaxPooling1D, ConvLSTM2D, Flatten, Add, LayerNormalization
from tensorflow.keras.utils import plot_model
import numpy as np

vocab_size = 10000
embedding_dim = 64
max_length = 100
hidden_units = 32
batch_size = 10
filters = 64
kernel_size = 5

conv_timesteps = 10
conv_rows = 32
conv_cols = 32
conv_channels = 1

print("--- Building Elman (SimpleRNN) Model ---")
model_rnn = Sequential(name="Elman_SimpleRNN_Model")
model_rnn.add(Embedding(input_dim=vocab_size,
                        output_dim=embedding_dim,
                        input_length=max_length))
model_rnn.add(SimpleRNN(units=hidden_units))
model_rnn.add(Dense(1, activation='sigmoid'))

model_rnn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_rnn.build(input_shape=(None, max_length))

print(f"This model uses the 'SimpleRNN' layer.\n")
model_rnn.summary()

plot_model(model_rnn, to_file='rnn_model_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved RNN diagram to 'rnn_model_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Building Vanilla (Single-Layer) LSTM Model ---")
model_lstm_single = Sequential(name="Vanilla_LSTM_Model")
model_lstm_single.add(Embedding(input_dim=vocab_size,
                         output_dim=embedding_dim,
                         input_length=max_length))
model_lstm_single.add(LSTM(units=hidden_units))
model_lstm_single.add(Dense(1, activation='sigmoid'))

model_lstm_single.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_lstm_single.build(input_shape=(None, max_length))

print(f"This model uses one 'LSTM' layer.\n")
model_lstm_single.summary()

plot_model(model_lstm_single, to_file='lstm_single_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved Single-Layer LSTM diagram to 'lstm_single_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Building Stacked (Deep) LSTM Model ---")
model_lstm_stacked = Sequential(name="Stacked_LSTM_Model")

model_lstm_stacked.add(Embedding(input_dim=vocab_size,
                                 output_dim=embedding_dim,
                                 input_length=max_length))

model_lstm_stacked.add(LSTM(units=hidden_units, return_sequences=True))
model_lstm_stacked.add(LSTM(units=hidden_units))
model_lstm_stacked.add(Dense(1, activation='sigmoid'))

model_lstm_stacked.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_lstm_stacked.build(input_shape=(None, max_length))

print(f"This model uses two stacked 'LSTM' layers.\n")
model_lstm_stacked.summary()

plot_model(model_lstm_stacked, to_file='lstm_stacked_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved Stacked LSTM diagram to 'lstm_stacked_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Building Bidirectional LSTM (BiLSTM) Model ---")
model_bilstm = Sequential(name="Bidirectional_LSTM_Model")

model_bilstm.add(Embedding(input_dim=vocab_size,
                           output_dim=embedding_dim,
                           input_length=max_length))

model_bilstm.add(Bidirectional(LSTM(units=hidden_units)))
model_bilstm.add(Dense(1, activation='sigmoid'))

model_bilstm.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model_bilstm.build(input_shape=(None, max_length))

print(f"This model uses one 'Bidirectional(LSTM)' layer.\n")
model_bilstm.summary()

plot_model(model_bilstm, to_file='bilstm_model_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved BiLSTM diagram to 'bilstm_model_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Building Attention-based LSTM Model ---")

attn_inputs = Input(shape=(max_length,), name="Attention_Input")
attn_emb = Embedding(input_dim=vocab_size,
                     output_dim=embedding_dim,
                     input_length=max_length,
                     name="Attention_Embedding")(attn_inputs)

attn_lstm_out = Bidirectional(LSTM(hidden_units, return_sequences=True), name="Bidirectional_LSTM")(attn_emb)
attn_self_attention = Attention(name="Self_Attention")([attn_lstm_out, attn_lstm_out])
attn_context_vector = GlobalAveragePooling1D(name="Global_Pooling")(attn_self_attention)
attn_outputs = Dense(1, activation='sigmoid', name="Output")(attn_context_vector)
model_attention = Model(inputs=attn_inputs, outputs=attn_outputs, name="Attention_LSTM_Model")

model_attention.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print(f"This model uses BiLSTM + Attention (Functional API).\n")
model_attention.summary()

plot_model(model_attention, to_file='attention_lstm_model_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved Attention-LSTM diagram to 'attention_lstm_model_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Building CNN-LSTM (Hybrid) Model ---")

cnn_lstm_inputs = Input(shape=(max_length,), name="CNN_LSTM_Input")
cnn_lstm_emb = Embedding(input_dim=vocab_size,
                         output_dim=embedding_dim,
                         input_length=max_length,
                         name="CNN_LSTM_Embedding")(cnn_lstm_inputs)
cnn_out = Conv1D(filters=filters, kernel_size=kernel_size, activation='relu', name="Conv1D_Layer")(cnn_lstm_emb)
cnn_out = MaxPooling1D(pool_size=4, name="MaxPooling1D_Layer")(cnn_out)
lstm_out = LSTM(units=hidden_units, name="LSTM_Layer")(cnn_out)
cnn_lstm_outputs = Dense(1, activation='sigmoid', name="Output")(lstm_out)
model_cnn_lstm = Model(inputs=cnn_lstm_inputs, outputs=cnn_lstm_outputs, name="CNN_LSTM_Model")

model_cnn_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print(f"This model uses Conv1D + LSTM (Functional API).\n")
model_cnn_lstm.summary()

plot_model(model_cnn_lstm, to_file='cnn_lstm_model_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved CNN-LSTM diagram to 'cnn_lstm_model_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Building Residual LSTM (ResLSTM) Model ---")

res_inputs = Input(shape=(max_length,), name="Residual_Input")
res_emb = Embedding(input_dim=vocab_size,
                    output_dim=embedding_dim,
                    input_length=max_length,
                    name="Residual_Embedding")(res_inputs)

res_lstm_out = LSTM(hidden_units, return_sequences=True, name="Res_LSTM")(res_emb)

shortcut = Conv1D(hidden_units, 1, padding='same', name="Shortcut_Conv1D")(res_emb)

res_add = Add(name="Add_Shortcut")([res_lstm_out, shortcut])
res_add = LayerNormalization(name="LayerNorm")(res_add)

res_pool = GlobalAveragePooling1D(name="Global_Pooling")(res_add)
res_outputs = Dense(1, activation='sigmoid', name="Output")(res_pool)

model_res_lstm = Model(inputs=res_inputs, outputs=res_outputs, name="Residual_LSTM_Model")

model_res_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print(f"This model uses a Residual Connection (Functional API).\n")
model_res_lstm.summary()

plot_model(model_res_lstm, to_file='res_lstm_model_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved Res-LSTM diagram to 'res_lstm_model_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Building ConvLSTM2D Model ---")

conv_lstm_inputs = Input(shape=(conv_timesteps, conv_rows, conv_cols, conv_channels), name="ConvLSTM_Input")
conv_lstm_out = ConvLSTM2D(filters=filters,
                             kernel_size=(3, 3),
                             padding='same',
                             name="ConvLSTM2D_Layer")(conv_lstm_inputs)
conv_lstm_out = Flatten(name="Flatten_Layer")(conv_lstm_out)
conv_lstm_outputs = Dense(1, activation='sigmoid', name="Output")(conv_lstm_out)
model_conv_lstm = Model(inputs=conv_lstm_inputs, outputs=conv_lstm_outputs, name="ConvLSTM2D_Model")

model_conv_lstm.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print(f"This model uses ConvLSTM2D (Functional API).\n")
model_conv_lstm.summary()

plot_model(model_conv_lstm, to_file='conv_lstm_model_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved ConvLSTM2D diagram to 'conv_lstm_model_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Building Encoder-Decoder LSTM (Seq2Seq) Model ---")

encoder_inputs = Input(shape=(max_length,), name="Encoder_Input")
enc_emb = Embedding(input_dim=vocab_size,
                    output_dim=embedding_dim,
                    input_length=max_length,
                    name="Encoder_Embedding")(encoder_inputs)
encoder_lstm = LSTM(hidden_units, return_state=True, name="Encoder_LSTM")
encoder_outputs, state_h, state_c = encoder_lstm(enc_emb)
encoder_states = [state_h, state_c]

decoder_inputs = Input(shape=(max_length,), name="Decoder_Input")
dec_emb_layer = Embedding(input_dim=vocab_size,
                          output_dim=embedding_dim,
                          input_length=max_length,
                          name="Decoder_Embedding")
dec_emb = dec_emb_layer(decoder_inputs)
decoder_lstm = LSTM(hidden_units, return_sequences=True, return_state=False, name="Decoder_LSTM")
decoder_outputs = decoder_lstm(dec_emb, initial_state=encoder_states)
decoder_dense = TimeDistributed(Dense(1, activation='sigmoid'), name="Time_Distributed_Dense")
decoder_outputs = decoder_dense(decoder_outputs)

model_seq2seq = Model([encoder_inputs, decoder_inputs], decoder_outputs, name="Encoder_Decoder_Model")

model_seq2seq.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print(f"This model uses the Functional API for Seq2Seq.\n")
model_seq2seq.summary()

plot_model(model_seq2seq, to_file='seq2seq_model_diagram.png', show_shapes=True, show_layer_names=True)
print("\nSaved Seq2Seq diagram to 'seq2seq_model_diagram.png'")

print("\n" + "="*40 + "\n")

print("--- Creating Dummy Data for Models 1-7 (Classification) ---")
dummy_x = np.random.randint(0, vocab_size, size=(batch_size, max_length))
dummy_y = np.random.randint(0, 2, size=(batch_size, 1))
print(f"Shape of standard input data (dummy_x): {dummy_x.shape}")
print(f"Shape of standard output data (dummy_y): {dummy_y.shape}")

print("\n" + "="*40 + "\n")

print("--- Creating Dummy Data for ConvLSTM2D Model 8 ---")
conv_dummy_x = np.random.rand(batch_size, conv_timesteps, conv_rows, conv_cols, conv_channels)
conv_dummy_y = np.random.randint(0, 2, size=(batch_size, 1))
print(f"Shape of ConvLSTM input data: {conv_dummy_x.shape}")
print(f"Shape of ConvLSTM output data: {conv_dummy_y.shape}")

print("\n" + "="*40 + "\n")

print("--- Creating Dummy Data for Seq2Seq Model 9 ---")
encoder_input_data = np.random.randint(0, vocab_size, size=(batch_size, max_length))
decoder_input_data = np.random.randint(0, vocab_size, size=(batch_size, max_length))
decoder_target_data = np.random.randint(0, 2, size=(batch_size, max_length, 1))
print(f"Shape of Seq2Seq encoder input: {encoder_input_data.shape}")
print(f"Shape of Seq2Seq decoder input: {decoder_input_data.shape}")
print(f"Shape of Seq2Seq decoder output: {decoder_target_data.shape}")

print("\n" + "="*40 + "\n")

print("--- Training Elman (SimpleRNN) Model ---")
model_rnn.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size, verbose=1)
print("SimpleRNN training complete.")

print("\n" + "="*40 + "\n")

print("--- Training Vanilla (Single-Layer) LSTM Model ---")
model_lstm_single.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size, verbose=1)
print("Single-Layer LSTM training complete.")

print("\n" + "="*40 + "\n")

print("--- Training Stacked (Deep) LSTM Model ---")
model_lstm_stacked.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size, verbose=1)
print("Stacked LSTM training complete.")

print("\n" + "="*40 + "\n")

print("--- Training Bidirectional LSTM (BiLSTM) Model ---")
model_bilstm.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size, verbose=1)
print("BiLSTM training complete.")

print("\n" + "="*40 + "\n")

print("--- Training Attention-based LSTM Model ---")
model_attention.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size, verbose=1)
print("Attention-LSTM training complete.")

print("\n" + "="*40 + "\n")

print("--- Training CNN-LSTM (Hybrid) Model ---")
model_cnn_lstm.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size, verbose=1)
print("CNN-LSTM training complete.")

print("\n" + "="*40 + "\n")

print("--- Training Residual LSTM (ResLSTM) Model ---")
model_res_lstm.fit(dummy_x, dummy_y, epochs=3, batch_size=batch_size, verbose=1)
print("Res-LSTM training complete.")

print("\n" + "="*40 + "\n")

print("--- Training ConvLSTM2D Model ---")
model_conv_lstm.fit(conv_dummy_x, conv_dummy_y, epochs=3, batch_size=batch_size, verbose=1)
print("ConvLSTM2D training complete.")

print("\n" + "="*40 + "\n")

print("--- Training Encoder-Decoder LSTM (Seq2Seq) Model ---")
model_seq2seq.fit([encoder_input_data, decoder_input_data], decoder_target_data,
                  epochs=3,
                  batch_size=batch_size,
                  verbose=1)
print("Seq2Seq training complete.")
```
