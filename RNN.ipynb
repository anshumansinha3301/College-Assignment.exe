# üß† Recurrent Neural Network (RNN)


---

## üìò Introduction

A **Recurrent Neural Network (RNN)** is a type of **artificial neural network** designed for processing **sequential data** ‚Äî where the order of data points matters.

Unlike feedforward networks, RNNs have connections that form **directed cycles**, allowing information to persist over time.  
This makes them suitable for tasks like **time series prediction**, **speech recognition**, and **natural language processing**.

---

## ‚öôÔ∏è Architecture Overview

An RNN takes an input sequence \( X = (x_1, x_2, ..., x_T) \) and processes it step by step, maintaining a **hidden state** \( h_t \) that captures past information.


## üîÅ Flow of Information

1. The RNN reads one element of the sequence at a time.  
2. At each step, it combines the **current input** with the **previous hidden state**.  
3. This updated hidden state carries information from all previous steps.  
4. The final output (or sequence of outputs) is computed based on these hidden states.

---

## üß© Key Characteristics

- **Sequential processing**: Input order affects output.  
- **Shared weights**: Same weights are used across all time steps.  
- **Temporal memory**: Keeps track of previous context via hidden state.  
- **Backpropagation Through Time (BPTT)**: Special training method for RNNs that unrolls the network across time steps.

---

## üìä Types of RNN Architectures

| Type | Description | Example Use |
|:------|:-------------|:-------------|
| **One-to-One** | Standard NN (no sequence) | Image classification |
| **One-to-Many** | One input ‚Üí sequence output | Image captioning |
| **Many-to-One** | Sequence input ‚Üí single output | Sentiment analysis |
| **Many-to-Many** | Sequence input ‚Üí sequence output | Machine translation |

---

## üßÆ Variants of RNN

| Variant | Full Form | Key Idea |
|:----------|:------------|:-----------|
| **LSTM** | Long Short-Term Memory | Adds memory cells & gates to prevent vanishing gradient |
| **GRU** | Gated Recurrent Unit | Simplified LSTM with fewer gates |
| **Bidirectional RNN** | Processes sequence from both directions | Improves context understanding |
| **Stacked RNN** | Multiple RNN layers | Learns deeper temporal patterns |

---

## üí° Applications

| Domain | Example Applications |
|:-------|:----------------------|
| **Natural Language Processing** | Language modeling, text classification, NER, machine translation |
| **Speech Recognition** | Voice commands, phoneme recognition |
| **Time Series Forecasting** | Stock prediction, weather forecasting, anomaly detection |
| **Healthcare** | ECG/EEG analysis, patient monitoring |
| **Finance** | Fraud detection, risk modeling |
| **Control Systems** | Sequence-based control or robotic motion |

---

## üß† Advantages

‚úÖ Handles variable-length sequential data  
‚úÖ Remembers past information (context awareness)  
‚úÖ Works well for time-dependent problems  
‚úÖ Can be stacked or combined with CNNs or Attention layers  
‚úÖ Lightweight compared to Transformers for small datasets  

---

## ‚ö†Ô∏è Limitations

‚ùå Difficult to learn long-term dependencies (vanishing gradient problem)  
‚ùå Sequential nature ‚Üí slow training (not parallelizable)  
‚ùå Prone to exploding gradients  
‚ùå Memory limited to short contexts  
‚ùå Often replaced by LSTMs, GRUs, or Transformers in modern NLP  

---

## üß© Improvements to Vanilla RNN

1. **Gradient Clipping** ‚Äì prevents exploding gradients.  
2. **LSTM / GRU** ‚Äì solves vanishing gradient issue.  
3. **Bidirectional RNNs** ‚Äì incorporate past & future context.  
4. **Attention Mechanisms** ‚Äì allow focus on relevant parts of the input sequence.  
5. **Transformers** ‚Äì remove recurrence entirely (faster & more accurate for long sequences).

---

## üî¨ Training: Backpropagation Through Time (BPTT)

RNNs are trained using **Backpropagation Through Time**, which:
- Unrolls the RNN across *T* time steps.
- Applies standard backpropagation to compute gradients.
- Accumulates gradients across time to update weights.

### Problems:
- **Vanishing gradients**: early timesteps have almost no effect.  
- **Exploding gradients**: gradients grow exponentially large.

**Solutions:**  
- Use **LSTMs or GRUs**  
- Apply **gradient clipping**  
- Use **layer normalization** or **residual connections**

---

## üßæ Activation Functions Used

| Function | Equation | Use |
|:----------|:-----------|:----|
| **Tanh** | \( \tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} \) | Keeps hidden states in (-1, 1) |
| **ReLU** | \( \max(0, x) \) | Helps reduce vanishing gradients |
| **Sigmoid** | \( \sigma(x) = \frac{1}{1 + e^{-x}} \) | Used in gates (LSTM/GRU) |

---

## üß≠ Comparison Table

| Model | Handles Long-Term Dependencies | Speed | Memory | Common Use |
|:-------|:------------------------------|:--------|:---------|:-------------|

| **Transformer** | ‚úÖ Excellent | Very Fast (parallel) | High | Modern NLP |

---

## ‚öôÔ∏è Tips for Practitioners

- Always **normalize input sequences** (helps stability).  
- Use **gradient clipping** (e.g., `torch.nn.utils.clip_grad_norm_`).  
- **Batch-first** inputs are easier to debug.  
- Combine RNNs with **CNNs** for spatial-temporal data (video/audio).  
- Use **Bidirectional RNN** if future context is available.  
- Start simple (RNN ‚Üí GRU ‚Üí LSTM ‚Üí Transformer).

# ELMAN NETWORK (VANILLA RNN)
# Sentiment Analysis on IMDB Movie Reviews using Elman Recurrent Neural Network (Vanilla RNN)

```python
# ============================================================
# üß† Elman Network (Vanilla RNN) on IMDB Dataset
# Visualization (Plotly)
# ============================================================

import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, SimpleRNN, Dense, Dropout
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
import plotly.graph_objects as go
import matplotlib.pyplot as plt

# -------------------------------
# 1Ô∏è‚É£ Load & Preprocess Dataset
# -------------------------------
max_features = 5000
maxlen = 200
embedding_dim = 64
hidden_units = 64
batch_size = 64
epochs = 6

print("üì• Loading IMDB dataset...")
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=max_features)

print(f"Training samples: {len(X_train)}, Test samples: {len(X_test)}")

X_train = sequence.pad_sequences(X_train, maxlen=maxlen)
X_test = sequence.pad_sequences(X_test, maxlen=maxlen)

# -------------------------------
# 2Ô∏è‚É£ Define Vanilla RNN Model
# -------------------------------
model = Sequential([
    Embedding(input_dim=max_features, output_dim=embedding_dim, input_length=maxlen),
    SimpleRNN(hidden_units, activation='tanh', dropout=0.2, recurrent_dropout=0.2),
    Dense(64, activation='relu'),
    Dropout(0.3),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])

# -------------------------------
# 3Ô∏è‚É£ Model Summary & Parameters
# -------------------------------
model.summary()

# -------------------------------
# 4Ô∏è‚É£ Train Model
# -------------------------------
history = model.fit(
    X_train, y_train,
    epochs=epochs,
    batch_size=batch_size,
    validation_split=0.2,
    verbose=1
)

# -------------------------------
# 5Ô∏è‚É£ Evaluate Model
# -------------------------------
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"\n‚úÖ Test Accuracy: {acc*100:.2f}%")

# -------------------------------
# 6Ô∏è‚É£ Plot Accuracy & Loss (Matplotlib)
# -------------------------------
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title('Accuracy')
plt.legend()

plt.subplot(1,2,2)
plt.plot(history.history['loss'], label='Train Loss')
plt.plot(history.history['val_loss'], label='Val Loss')
plt.title('Loss')
plt.legend()
plt.tight_layout()
plt.show()

# -------------------------------
# 7Ô∏è‚É£ Interactive Architecture Diagram (Plotly)
# -------------------------------
layers = [layer.__class__.__name__ for layer in model.layers]
params = [layer.count_params() for layer in model.layers]

fig = go.Figure(data=[go.Bar(
    x=layers,
    y=params,
    marker_color='skyblue',
    text=params,
    textposition='auto'
)])

fig.update_layout(
    title="üß± Elman (SimpleRNN) Architecture ‚Äî Layer Parameter Count",
    xaxis_title="Layers",
    yaxis_title="Parameters",
    template="plotly_dark"
)

fig.show()
```

# Jordan Recurrent Neural Network (RNN)

# Jordan Network Implementation in TensorFlow

```python
import tensorflow as tf
from tensorflow.keras.layers import Layer, RNN, Dense, Input, Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from tensorflow.keras.utils import plot_model
from IPython.display import Image, display

print(f"Using TensorFlow Version: {tf.__version__}")

# -----------------------------------------------------------------------------
# 1. ARCHITECTURE DETAILS & VISUALIZATION
# -----------------------------------------------------------------------------
"""
=================================================
Jordan Recurrent Neural Network (RNN)
=================================================

A Jordan Network (1990) is a type of simple recurrent neural network. Its
distinguishing feature is its "context layer" or "state."

-   **Elman Network (Standard SimpleRNN):** Feeds the HIDDEN layer's
    output from the previous timestep (h_t-1) back as input to the
    hidden layer at the current timestep (h_t).

-   **Jordan Network:** Feeds the OUTPUT layer's output from the
    previous timestep (y_t-1) back as input to the hidden layer
    at the current timestep (h_t).

This means the "memory" of the network is based on what it actually
predicted in the last step, not just its internal hidden activation.

---
Data Flow Visualization (at timestep 't'):
---

                    [ Input (x_t) ]
                           |
                           v
     +-----------------[ H_t ]-----------------+
     | (Hidden Layer)                          |
     |                                         |
     |   h_t = activation( (W * x_t) +          |
     |                     (U * y_t-1) + b )   |
     |                                         |
     +-------------------v---------------------+
                           |
                           v
                     [ Output (y_t) ] ----+
                                          |
     (Loop for next timestep)             |
     y_t becomes y_t-1 for (t+1)          |
     <------------------------------------+

To implement this, we create a custom Keras RNN 'Cell'.
The 'state' of our cell will be the previous output (y_t-1).
"""


# -----------------------------------------------------------------------------
# 2. CODE: CUSTOM JORDAN RNN CELL
# -----------------------------------------------------------------------------

class JordanCell(Layer):
    """
    Custom RNN cell implementing the Jordan Network logic.

    Inherits from tf.keras.layers.Layer.

    The 'state' for this cell is the 'output' from the previous timestep.
    """
    def __init__(self, units, activation='tanh', **kwargs):
        super(JordanCell, self).__init__(**kwargs)
        self.units = units
        self.activation = tf.keras.activations.get(activation)

        self.state_size = self.units

    def build(self, input_shape):
        """
        Create the weights for the cell. This method is called by Keras
        automatically the first time the layer is used.
        """

        self.kernel = self.add_weight(
            shape=(input_shape[-1], self.units),
            initializer='glorot_uniform',
            name='kernel'
        )

        self.recurrent_kernel = self.add_weight(
            shape=(self.units, self.units),
            initializer='orthogonal',
            name='recurrent_kernel'
        )

        self.bias = self.add_weight(
            shape=(self.units,),
            initializer='zeros',
            name='bias'
        )

        self.built = True

    def call(self, inputs, states):
        """
        The core logic of the cell, processed at each timestep.

        Args:
            inputs (tensor): The input at the current timestep (x_t).
            states (list): A list containing the state from the
                           previous timestep [y_t-1].

        Returns:
            A tuple (output, new_state):
            - output (tensor): The output for this timestep (y_t).
            - new_state (list): A list containing the new state [y_t].
        """

        prev_output = states[0]

        input_part = tf.matmul(inputs, self.kernel)

        state_part = tf.matmul(prev_output, self.recurrent_kernel)

        output = self.activation(input_part + state_part + self.bias)

        new_state = [output]

        return output, new_state

    def get_config(self):
        """For model saving and loading."""
        config = super(JordanCell, self).get_config()
        config.update({
            'units': self.units,
            'activation': tf.keras.activations.serialize(self.activation)
        })
        return config

print("Custom 'JordanCell' class defined successfully.")


# -----------------------------------------------------------------------------
# 3. CODE: MODEL BUILDING & SUMMARY
# -----------------------------------------------------------------------------

VOCAB_SIZE = 10000
EMBEDDING_DIM = 16
MAX_LEN = 120
N_UNITS = 32

print("\n--- Building Model ---")

jordan_cell_instance = JordanCell(N_UNITS)

model = Sequential([
    # Input layer specifies the shape of the sequences (now MAX_LEN)
    Input(shape=(MAX_LEN,)),

    # 1. Embedding Layer
    # Turns positive integers (word indices) into dense vectors.
    # Input shape: (batch_size, MAX_LEN)
    # Output shape: (batch_size, MAX_LEN, EMBEDDING_DIM)
    Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM),

    # 2. Jordan RNN Layer
    # The 'RNN' layer acts as a wrapper.
    # It takes our custom 'cell' and unrolls it for the full sequence.
    # It automatically receives the (batch_size, MAX_LEN, EMBEDDING_DIM)
    # output from the Embedding layer.
    RNN(jordan_cell_instance, return_sequences=False),

    # 3. Final Output Layer
    # A final Dense layer to produce the single classification output
    # Use 'sigmoid' activation for binary (0 or 1) classification.
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("\n--- Model Summary ---")
model.summary()

print("\n--- Model Architecture Diagram ---")
try:
    plot_model(
        model,
        to_file='jordan_model_diagram.png',
        show_shapes=True,
        show_layer_names=True,
        show_layer_activations=True,
        dpi=70
    )

    print("Displaying model architecture diagram (saved as 'jordan_model_diagram.png'):")
    display(Image(filename='jordan_model_diagram.png'))
except ImportError as e:
    print(f"\nModel visualization failed. This often requires 'pydot' and 'graphviz'.")
    print("In Google Colab, you can install them by running:")
    print("!pip install pydot graphviz")
    print(f"Error: {e}")


# -----------------------------------------------------------------------------
# 4. CODE: MODEL TRAINING
# -----------------------------------------------------------------------------

print("\n--- Loading IMDB Built-in Dataset ---")

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=VOCAB_SIZE)

print(f"Loaded {len(X_train)} training sequences and {len(X_test)} test sequences.")
print("Sequences have varying lengths.")

print("\n--- Preprocessing Data (Padding) ---")
X_train = pad_sequences(X_train, maxlen=MAX_LEN, padding='post', truncating='post')
X_test = pad_sequences(X_test, maxlen=MAX_LEN, padding='post', truncating='post')

print(f"Padded training data (X) shape: {X_train.shape}")
print(f"Padded test data (X) shape: {X_test.shape}")
print(f"Training labels (y) shape: {y_train.shape}")

print("\n--- Training the Jordan Network ---")
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_data=(X_test, y_test),
    verbose=1
)

print("Training complete.")


# -----------------------------------------------------------------------------
# 5. CODE: MODEL PREDICTION (TESTING)
# -----------------------------------------------------------------------------

print("\n--- Evaluating Model on Test Data ---")

results = model.evaluate(X_test, y_test, verbose=1)

print("\n--- Final Test Results ---")
print(f"Test Loss:     {results[0]:.4f}")
print(f"Test Accuracy: {results[1]:.4f} ({(results[1] * 100):.2f}%)")
```

# LSTM Implementation

```python
import tensorflow as tf
from tensorflow.keras.layers import Layer, RNN, Dense, Input, Embedding, LSTM
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from tensorflow.keras.utils import plot_model
from IPython.display import Image, display

print(f"Using TensorFlow Version: {tf.__version__}")

# -----------------------------------------------------------------------------
# 1. ARCHITECTURE DETAILS & VISUALIZATION
# -----------------------------------------------------------------------------
"""
=================================================
LSTM (Long Short-Term Memory)
=================================================
- **Inventors:** Hochreiter & Schmidhuber
- **Year:** 1997
- **Key Concept:** Introduced memory cells and gating mechanisms
  to solve the vanishing gradient problem in standard RNNs.

An LSTM cell is a complex unit with its own internal "cell state" (C_t)
that acts as a memory conveyor belt. It can add or remove information
from this state using regulated "gates":

1.  **Forget Gate (f_t):** Decides what information from the previous
    cell state (C_t-1) to throw away.
2.  **Input Gate (i_t):** Decides which new information from the current
    input (x_t) and previous hidden state (h_t-1) to store in the
    current cell state (C_t).
3.  **Output Gate (o_t):** Decides what to output as the new hidden
    state (h_t), based on the new cell state (C_t).

---
Data Flow Visualization (at timestep 't'):
---

     [ Prev. Cell State (C_t-1) ]------+----(Forget Gate)----+
                                       |                     |
     [ Input (x_t) ] --+               |                     |
                       v               v                     v
     [ Prev. Hidden (h_t-1) ] --+---> [ GATES ] --(Input Gate)--> [ New Cell State (C_t) ]
                                |      (f, i, o)             |
                                |                            |
     (Loop for next timestep)     |                            v
     C_t and h_t become (t+1)     +----(Output Gate)--------> [ New Hidden (h_t) ]
     <-------------------------------------------------------(Also the output y_t)

Because LSTM is a standard, built-in layer in Keras, we do NOT
need to write a custom cell. We can just use `tf.keras.layers.LSTM`.
"""

print("LSTM architecture details loaded.")


# -----------------------------------------------------------------------------
# 2. CODE: MODEL BUILDING & SUMMARY
# -----------------------------------------------------------------------------

VOCAB_SIZE = 10000
EMBEDDING_DIM = 16
MAX_LEN = 120
N_UNITS = 32

print("\n--- Building Model ---")

model = Sequential([
    Input(shape=(MAX_LEN,)),

    # 1. Embedding Layer
    # Turns positive integers (word indices) into dense vectors.
    # Input shape: (batch_size, MAX_LEN)
    # Output shape: (batch_size, MAX_LEN, EMBEDDING_DIM)
    Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM),

    # 2. LSTM Layer
    # We simply use the built-in LSTM layer. Keras handles all the
    # complex gate and cell state logic internally.
    # return_sequences=False means we only get the output from the
    # *final* timestep, which is what we want for classification.
    LSTM(N_UNITS, return_sequences=False),

    # 3. Final Output Layer
    # A final Dense layer to produce the single classification output
    # Use 'sigmoid' activation for binary (0 or 1) classification.
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("\n--- Model Summary ---")
model.summary()

print("\n--- Model Architecture Diagram ---")
try:
    plot_model(
        model,
        to_file='lstm_model_diagram.png',
        show_shapes=True,
        show_layer_names=True,
        show_layer_activations=True,
        dpi=70
    )

    print("Displaying model architecture diagram (saved as 'lstm_model_diagram.png'):")
    display(Image(filename='lstm_model_diagram.png'))
except ImportError as e:
    print(f"\nModel visualization failed. This often requires 'pydot' and 'graphviz'.")
    print("In Google Colab, you can install them by running:")
    print("!pip install pydot graphviz")
    print(f"Error: {e}")


# -----------------------------------------------------------------------------
# 3. CODE: MODEL TRAINING
# -----------------------------------------------------------------------------

print("\n--- Loading IMDB Built-in Dataset ---")
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=VOCAB_SIZE)

print(f"Loaded {len(X_train)} training sequences and {len(X_test)} test sequences.")
print("Sequences have varying lengths.")

print("\n--- Preprocessing Data (Padding) ---")
X_train = pad_sequences(X_train, maxlen=MAX_LEN, padding='post', truncating='post')
X_test = pad_sequences(X_test, maxlen=MAX_LEN, padding='post', truncating='post')

print(f"Padded training data (X) shape: {X_train.shape}")
print(f"Padded test data (X) shape: {X_test.shape}")
print(f"Training labels (y) shape: {y_train.shape}")

print("\n--- Training the LSTM Network ---")
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_data=(X_test, y_test),
    verbose=1
)

print("Training complete.")


# -----------------------------------------------------------------------------
# 4. CODE: MODEL PREDICTION (TESTING)
# -----------------------------------------------------------------------------

print("\n--- Evaluating Model on Test Data ---")

results = model.evaluate(X_test, y_test, verbose=1)

print("\n--- Final Test Results ---")
print(f"Test Loss:     {results[0]:.4f}")
print(f"Test Accuracy: {results[1]:.4f} ({(results[1] * 100):.2f}%)")
```

# GRU Implementation in TensorFlow

```python
import tensorflow as tf
from tensorflow.keras.layers import Layer, RNN, Dense, Input, Embedding, GRU
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from tensorflow.keras.utils import plot_model
from IPython.display import Image, display

print(f"Using TensorFlow Version: {tf.__version__}")

# -----------------------------------------------------------------------------
# 1. ARCHITECTURE DETAILS & VISUALIZATION
# -----------------------------------------------------------------------------
"""
=================================================
GRU (Gated Recurrent Unit)
=================================================
- **Inventors:** Cho et al.
- **Year:** 2014
- **Key Concept:** Simplified LSTM with fewer gates, making it
  faster to compute and equally effective in many cases.

A GRU merges the "cell state" (C_t) and "hidden state" (h_t)
from the LSTM into a single hidden state (h_t). It also combines
the "forget" and "input" gates into a single "update" gate.

It uses two main gates:

1.  **Update Gate (z_t):** Decides how much of the previous hidden
    state (h_t-1) to keep versus how much of the "candidate"
    new state to add. This is similar to LSTM's forget/input gates.
2.  **Reset Gate (r_t):** Decides how much of the previous hidden
    state (h_t-1) to forget before calculating the candidate state.

---
Data Flow Visualization (at timestep 't'):
---

                                 [ Candidate State (h'_t) ]
                                          ^
                                          |
     [ Input (x_t) ] --+                  |
                       v                  |
     [ Prev. Hidden (h_t-1) ] --+---> [ GATES ] ----+
                                |      (z, r)      |
                                |                  |
     (Loop for next timestep)     | (Reset Gate 'r') |
     h_t becomes h_t-1            |                  |
     <----------------------------+---(Update Gate 'z')--> [ New Hidden (h_t) ]
                                                       (Also the output y_t)

Like LSTM, GRU is a standard, built-in layer in Keras.
We can just use `tf.keras.layers.GRU`.
"""

print("GRU architecture details loaded.")


# -----------------------------------------------------------------------------
# 2. CODE: MODEL BUILDING & SUMMARY
# -----------------------------------------------------------------------------

VOCAB_SIZE = 10000
EMBEDDING_DIM = 16
MAX_LEN = 120
N_UNITS = 32

print("\n--- Building Model ---")

model = Sequential([
    Input(shape=(MAX_LEN,)),

    # 1. Embedding Layer
    # Turns positive integers (word indices) into dense vectors.
    # Input shape: (batch_size, MAX_LEN)
    # Output shape: (batch_size, MAX_LEN, EMBEDDING_DIM)
    Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM),

    # 2. GRU Layer
    # We simply use the built-in GRU layer.
    # It's a direct replacement for the LSTM layer.
    # return_sequences=False means we only get the output from the
    # *final* timestep, which is what we want for classification.
    GRU(N_UNITS, return_sequences=False),

    # 3. Final Output Layer
    # A final Dense layer to produce the single classification output
    # Use 'sigmoid' activation for binary (0 or 1) classification.
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

print("\n--- Model Summary ---")
model.summary()

print("\n--- Model Architecture Diagram ---")
try:
    plot_model(
        model,
        to_file='gru_model_diagram.png',
        show_shapes=True,
        show_layer_names=True,
        show_layer_activations=True,
        dpi=70
    )

    print("Displaying model architecture diagram (saved as 'gru_model_diagram.png'):")
    display(Image(filename='gru_model_diagram.png'))
except ImportError as e:
    print(f"\nModel visualization failed. This often requires 'pydot' and 'graphviz'.")
    print("In Google Colab, you can install them by running:")
    print("!pip install pydot graphviz")
    print(f"Error: {e}")


# -----------------------------------------------------------------------------
# 3. CODE: MODEL TRAINING
# -----------------------------------------------------------------------------

print("\n--- Loading IMDB Built-in Dataset ---")

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=VOCAB_SIZE)

print(f"Loaded {len(X_train)} training sequences and {len(X_test)} test sequences.")
print("Sequences have varying lengths.")

print("\n--- Preprocessing Data (Padding) ---")
X_train = pad_sequences(X_train, maxlen=MAX_LEN, padding='post', truncating='post')
X_test = pad_sequences(X_test, maxlen=MAX_LEN, padding='post', truncating='post')

print(f"Padded training data (X) shape: {X_train.shape}")
print(f"Padded test data (X) shape: {X_test.shape}")
print(f"Training labels (y) shape: {y_train.shape}")

print("\n--- Training the GRU Network ---")
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_data=(X_test, y_test),
    verbose=1
)

print("Training complete.")


# -----------------------------------------------------------------------------
# 4. CODE: MODEL PREDICTION (TESTING)
# -----------------------------------------------------------------------------

print("\n--- Evaluating Model on Test Data ---")

results = model.evaluate(X_test, y_test, verbose=1)

print("\n--- Final Test Results ---")
print(f"Test Loss:     {results[0]:.4f}")
print(f"Test Accuracy: {results[1]:.4f} ({(results[1] * 100):.2f}%)")
```

# Bidirectional RNN Implementation in TensorFlow

```python
import tensorflow as tf
from tensorflow.keras.layers import Layer, RNN, Dense, Input, Embedding, GRU, Bidirectional
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from tensorflow.keras.utils import plot_model
from IPython.display import Image, display

print(f"Using TensorFlow Version: {tf.__version__}")

# -----------------------------------------------------------------------------
# 1. ARCHITECTURE DETAILS & VISUALIZATION
# -----------------------------------------------------------------------------
"""
=================================================
Bidirectional RNN (BiRNN)
=================================================
- **Inventors:** Schuster & Paliwal
- **Year:** 1997
- **Key Concept:** Processes sequences in both forward and backward
  directions to provide richer context.

A standard RNN, LSTM, or GRU only processes the sequence in one
direction (forward). This means the prediction for a word at
timestep 't' only has information from words 0 to 't-1'.

A Bidirectional RNN fixes this. It consists of two separate RNNs
(e.g., two GRUs or two LSTMs):

1.  **Forward Layer:** Processes the sequence from start to end (t=0 to t=n).
2.  **Backward Layer:** Processes the sequence from end to start (t=n to t=0).

The outputs from both layers are then combined (e.g., concatenated)
at each timestep. This gives every timestep a complete context of
*both* past and future words in the sequence.

---
Data Flow Visualization (for a sequence of 3):
---

                 [x_0]       [x_1]       [x_2]
                  |           |           |
                  v           v           v
Forward RNN:   ->[h_f0]----->[h_f1]----->[h_f2]->
                  ^           ^           ^
                  |           |           |
Backward RNN:  <-[h_b0]<-----[h_b1]<-----[h_b2]<-
                  |           |           |
                  v           v           v
Output:      [y_0_comb]  [y_1_comb]  [y_2_comb]

In Keras, we don't build this manually. We use the
`tf.keras.layers.Bidirectional` wrapper, which handles
creating and combining the two RNN layers for us.
"""

print("Bidirectional RNN (BiRNN) architecture details loaded.")


# -----------------------------------------------------------------------------
# 2. CODE: MODEL BUILDING & SUMMARY
# -----------------------------------------------------------------------------

VOCAB_SIZE = 10000
EMBEDDING_DIM = 16
MAX_LEN = 120
N_UNITS = 32

print("\n--- Building Model ---")
model = Sequential([
    Input(shape=(MAX_LEN,)),

    Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM),

    Bidirectional(GRU(N_UNITS, return_sequences=False)),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print("\n--- Model Summary ---")
model.summary()

print("\n--- Model Architecture Diagram ---")
try:
    plot_model(
        model,
        to_file='birnn_model_diagram.png',
        show_shapes=True,
        show_layer_names=True,
        show_layer_activations=True,
        dpi=70
    )

    print("Displaying model architecture diagram (saved as 'birnn_model_diagram.png'):")
    display(Image(filename='birnn_model_diagram.png'))
except ImportError as e:
    print(f"\nModel visualization failed. This often requires 'pydot' and 'graphviz'.")
    print("In Google Colab, you can install them by running:")
    print("!pip install pydot graphviz")
    print(f"Error: {e}")


# -----------------------------------------------------------------------------
# 3. CODE: MODEL TRAINING
# -----------------------------------------------------------------------------

print("\n--- Loading IMDB Built-in Dataset ---")

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=VOCAB_SIZE)

print(f"Loaded {len(X_train)} training sequences and {len(X_test)} test sequences.")
print("Sequences have varying lengths.")

print("\n--- Preprocessing Data (Padding) ---")
X_train = pad_sequences(X_train, maxlen=MAX_LEN, padding='post', truncating='post')
X_test = pad_sequences(X_test, maxlen=MAX_LEN, padding='post', truncating='post')

print(f"Padded training data (X) shape: {X_train.shape}")
print(f"Padded test data (X) shape: {X_test.shape}")
print(f"Training labels (y) shape: {y_train.shape}")

print("\n--- Training the Bidirectional GRU Network ---")
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=64,
    validation_data=(X_test, y_test),
    verbose=1
)

print("Training complete.")


# -----------------------------------------------------------------------------
# 4. CODE: MODEL PREDICTION (TESTING)
# -----------------------------------------------------------------------------

print("\n--- Evaluating Model on Test Data ---")

results = model.evaluate(X_test, y_test, verbose=1)

print("\n--- Final Test Results ---")
print(f"Test Loss:     {results[0]:.4f}")
print(f"Test Accuracy: {results[1]:.4f} ({(results[1] * 100):.2f}%)")
```

# Seq2Seq (Encoder-Decoder) Implementation in TensorFlow

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from tensorflow.keras.utils import plot_model
from IPython.display import Image, display
import os
import zipfile
import string

print(f"Using TensorFlow Version: {tf.__version__}")

# -----------------------------------------------------------------------------
# 1. ARCHITECTURE DETAILS & VISUALIZATION
# -----------------------------------------------------------------------------
"""
=================================================
Encoder‚ÄìDecoder (Seq2Seq RNN)
=================================================
- **Inventors:** Sutskever, Vinyals & Le
- **Year:** 2014
- **Key Concept:** Uses two separate RNNs to map an input sequence
  (e.g., English) to a different output sequence (e.g., French).

The architecture consists of two main components:

1.  **The Encoder:**
    * An RNN (like an LSTM or GRU) that reads the entire input sequence
      (e.g., "Hello world").
    * Its goal is not to output anything, but to compress the
      information of the whole sequence into a final "state"
      vector (often called the "context" or "thought" vector).
    * This state (the final hidden state `h` and cell state `c`
      of the LSTM) is the *only* thing passed to the decoder.

2.  **The Decoder:**
    * A separate RNN that is initialized with the Encoder's
      final state vector.
    * Its job is to *generate* the output sequence word by word.
    * At timestep 't', it takes two inputs:
        1. The state from its *own* previous timestep (t-1).
        2. The *actual* word it generated at timestep (t-1).
    * It starts by being fed a special "<start>" token.
    * It stops when it generates a special "<end>" token.

---
Data Flow Visualization:
---

[ "Hello" ] -> [Encoder] -> [ "world" ] -> [Encoder] -> [Final State (h, c)]
                                                            |
                                                            v
[ "<start>" ] -> [Decoder (h, c)] -> [ "Bonjour" ] (y_0) -----+
                                      |                      |
                                      v                      v
                [ "Bonjour" ] -> [Decoder (h,c)] -> [ "le" ] (y_1) --...
                                      ...
                                      ... -> [ "<end>" ] (y_n)

This architecture *requires* the Keras Functional API, as we
are building a model with a non-sequential data flow.
"""

print("Encoder-Decoder (Seq2Seq) architecture details loaded.")


# -----------------------------------------------------------------------------
# 2. CODE: DATA PREPARATION
# -----------------------------------------------------------------------------


BATCH_SIZE = 64
N_UNITS = 256
NUM_SAMPLES = 10000
DATA_PATH = 'fra-eng/fra.txt'


print("\n--- Loading & Preparing Data ---")
if not os.path.exists('fra-eng'):
    print("Downloading dataset...")
    !wget -q http://storage.googleapis.com/download.tensorflow.org/data/fra-eng.zip
    with zipfile.ZipFile('fra-eng.zip', 'r') as zip_ref:
        zip_ref.extractall()
    print("Dataset downloaded and extracted.")

input_texts = []
target_texts = []
translator = str.maketrans('', '', string.punctuation)

with open(DATA_PATH, 'r', encoding='utf-8') as f:
    lines = f.read().split('\n')

for line in lines[: min(NUM_SAMPLES, len(lines) - 1)]:
    input_text, target_text, _ = line.split('\t')


    target_text = 'start_ ' + target_text.lower().translate(translator) + ' _end'
    input_text = input_text.lower().translate(translator)

    input_texts.append(input_text)
    target_texts.append(target_text)

print(f"Loaded {len(input_texts)} sentence pairs.")
print("Example:")
print(f"  Input: {input_texts[0]}")
print(f"  Target: {target_texts[0]}")


input_tokenizer = Tokenizer()
input_tokenizer.fit_on_texts(input_texts)
input_sequences = input_tokenizer.texts_to_sequences(input_texts)

target_tokenizer = Tokenizer(filters='')
target_tokenizer.fit_on_texts(target_texts)
target_sequences = target_tokenizer.texts_to_sequences(target_texts)

input_vocab_size = len(input_tokenizer.word_index) + 1
target_vocab_size = len(target_tokenizer.word_index) + 1

max_encoder_seq_len = max(len(s) for s in input_sequences)
max_decoder_seq_len = max(len(s) for s in target_sequences)

print(f"Input Vocab Size: {input_vocab_size}")
print(f"Target Vocab Size: {target_vocab_size}")
print(f"Max Input Length: {max_encoder_seq_len}")
print(f"Max Target Length: {max_decoder_seq_len}")


encoder_input_data = pad_sequences(input_sequences, maxlen=max_encoder_seq_len, padding='post')

decoder_input_data = pad_sequences(target_sequences, maxlen=max_decoder_seq_len, padding='post')


decoder_target_data = np.zeros_like(decoder_input_data)
for i, seq in enumerate(decoder_input_data):
    decoder_target_data[i, :-1] = seq[1:]


decoder_target_data = np.expand_dims(decoder_target_data, -1)

print(f"Shape of encoder_input_data: {encoder_input_data.shape}")
print(f"Shape of decoder_input_data: {decoder_input_data.shape}")
print(f"Shape of decoder_target_data: {decoder_target_data.shape}")


# -----------------------------------------------------------------------------
# 3. CODE: MODEL BUILDING (FUNCTIONAL API)
# -----------------------------------------------------------------------------
# We must use the Functional API to build this.
# We are actually building THREE models:
# 1. `training_model`: The full model for training.
# 2. `encoder_model`: The inference-only encoder.
# 3. `decoder_model`: The inference-only decoder.

print("\n--- Building Training Model ---")
encoder_inputs = Input(shape=(max_encoder_seq_len,), name='encoder_input')
enc_emb_layer = Embedding(input_vocab_size, N_UNITS, name='encoder_embedding')
enc_emb = enc_emb_layer(encoder_inputs)
encoder_lstm = LSTM(N_UNITS, return_state=True, name='encoder_lstm')
_, state_h, state_c = encoder_lstm(enc_emb)
encoder_states = [state_h, state_c]


decoder_inputs = Input(shape=(max_decoder_seq_len,), name='decoder_input')

dec_emb_layer = Embedding(target_vocab_size, N_UNITS, name='decoder_embedding')
dec_emb = dec_emb_layer(decoder_inputs)
decoder_lstm = LSTM(N_UNITS, return_sequences=True, return_state=True, name='decoder_lstm')
decoder_outputs, _, _ = decoder_lstm(
    dec_emb,
    initial_state=encoder_states
)
decoder_dense = Dense(target_vocab_size, activation='softmax', name='decoder_dense_output')
decoder_outputs = decoder_dense(decoder_outputs)

training_model = Model(
    [encoder_inputs, decoder_inputs],
    decoder_outputs,
    name='seq2seq_training_model'
)
training_model.compile(
    optimizer='rmsprop',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
print("\n--- Training Model Summary ---")
training_model.summary()

print("\n--- Model Architecture Diagram ---")
try:
    plot_model(
        training_model,
        to_file='seq2seq_model_diagram.png',
        show_shapes=True,
        show_layer_names=True,
        show_layer_activations=True,
        dpi=70
    )
    print("Displaying model architecture diagram (saved as 'seq2seq_model_diagram.png'):")
    display(Image(filename='seq2seq_model_diagram.png'))
except ImportError as e:
    print(f"\nModel visualization failed. {e}")
    print("!pip install pydot graphviz")


# -----------------------------------------------------------------------------
# 4. CODE: MODEL TRAINING
# -----------------------------------------------------------------------------

print("\n--- Training the Seq2Seq Network ---")

history = training_model.fit(
    [encoder_input_data, decoder_input_data],
    decoder_target_data,
    batch_size=BATCH_SIZE,
    epochs=20,
    validation_split=0.2,
    verbose=1
)

print("Training complete.")


# -----------------------------------------------------------------------------
# 5. CODE: MODEL PREDICTION (INFERENCE)
# -----------------------------------------------------------------------------

print("\n--- Building Inference Models ---")

encoder_model = Model(encoder_inputs, encoder_states, name='encoder_inference')
decoder_state_input_h = Input(shape=(N_UNITS,), name='decoder_state_h')
decoder_state_input_c = Input(shape=(N_UNITS,), name='decoder_state_c')
decoder_states_inputs = [decoder_state_input_h, decoder_state_input_c]

decoder_single_input = Input(shape=(1,), name='decoder_single_token_input')
dec_emb_inf = dec_emb_layer(decoder_single_input)
decoder_outputs_inf, state_h_inf, state_c_inf = decoder_lstm(
    dec_emb_inf, initial_state=decoder_states_inputs
)
decoder_states_inf = [state_h_inf, state_c_inf]
decoder_outputs_inf = decoder_dense(decoder_outputs_inf)
decoder_model = Model(
    [decoder_single_input] + decoder_states_inputs,
    [decoder_outputs_inf] + decoder_states_inf,
    name='decoder_inference'
)

print("Inference model summaries:")
encoder_model.summary(line_length=100)
decoder_model.summary(line_length=100)
reverse_input_word_index = dict((i, word) for word, i in input_tokenizer.word_index.items())
reverse_target_word_index = dict((i, word) for word, i in target_tokenizer.word_index.items())
target_start_token_index = target_tokenizer.word_index['start_']
target_end_token_index = target_tokenizer.word_index['_end']

def decode_sequence(input_seq):
    states_value = encoder_model.predict(input_seq, verbose=0)
    target_seq = np.zeros((1, 1))
    target_seq[0, 0] = target_start_token_index

    decoded_sentence = []
    stop_condition = False

    while not stop_condition:
        output_tokens, h, c = decoder_model.predict(
            [target_seq] + states_value, verbose=0
        )
        sampled_token_index = np.argmax(output_tokens[0, -1, :])
        sampled_word = reverse_target_word_index.get(sampled_token_index, '?')
        if (sampled_word == '_end' or
            len(decoded_sentence) > max_decoder_seq_len):
            stop_condition = True
        else:
            decoded_sentence.append(sampled_word)
        target_seq[0, 0] = sampled_token_index
        states_value = [h, c]
    return ' '.join(decoded_sentence)
print("\n--- Testing Inference ---")
for i in [10, 20, 30, 40, 50]:
    input_seq = encoder_input_data[i: i + 1]
    decoded_sentence = decode_sequence(input_seq)

    print('-')
    print('Input sentence:', input_texts[i])
    print('Expected sentence:', target_texts[i].replace('start_', '').replace('_end', '').strip())
    print('Decoded sentence:', decoded_sentence)
```

# Attention-based RNN (Seq2Seq) Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, LSTM, Dense, Bidirectional, Attention, GlobalAveragePooling1D
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from tensorflow.keras.utils import plot_model, get_file
from IPython.display import Image, display
import os
import string
from tensorflow.keras.datasets import imdb

print(f"Using TensorFlow Version: {tf.__version__}")

# -----------------------------------------------------------------------------
# 1. ARCHITECTURE DETAILS & VISUALIZATION
# -----------------------------------------------------------------------------
"""
=================================================
Attention-based RNN (for Classification)
=================================================
- **Inventors:** Bahdanau, Cho, & Bengio (concept), adapted for classification.
- **Year:** 2015
- **Key Concept:** Using an Attention layer to let the model weigh the
  importance of different parts of an input sequence for a
  single classification decision.

Instead of Seq2Seq (translation), we will use Attention for
classification. This avoids needing external files and uses
the built-in IMDB dataset.

The "bottleneck" problem still exists: in a normal RNN, only
the *final* hidden state is used for classification. This
means the model might forget important words from the beginning
of a long review.

**Attention** fixes this:

1.  **RNN Layer:** A Bidirectional RNN (like LSTM or GRU) processes
    the whole sequence and outputs a hidden state for *every* word.
    (e.D., `outputs` = [h_1, h_2, ..., h_n])

2.  **Attention Layer:** This layer is fed the entire sequence of
    RNN outputs. It learns to assign "importance scores" to each
    word's hidden state.

3.  **Context Vector:** It creates a single "context vector" by
    taking a weighted average of all the hidden states, based
    on the scores. This vector "focuses" on the most important
    parts of the review (e.g., "amazing", "awful").

4.  **Classifier:** This single, focused context vector is fed
    to the final `Dense` layer to make the positive/negative
    prediction.

---
Data Flow Visualization:
---
[ "This", "movie", "was", "amazing" ]
   |       |       |         |
   v       v       v         v
[ Embedding Layer ]
   |       |       |         |
   v       v       v         v
[ Bidirectional LSTM (outputs full sequence) ]
   |       |       |         |
   v       v       v         v
[  Attention Layer (calculates scores)  ]
   |
   v
[ Single Context Vector (weighted average) ]
   |
   v
[ Dense(1, activation='sigmoid') ]
   |
   v
[ "Positive" (0.98) ]
"""

print("Attention-for-Classification architecture details loaded.")


# -----------------------------------------------------------------------------
# 2. CODE: DATA PREPARATION (Using IMDB)
# -----------------------------------------------------------------------------

VOCAB_SIZE = 10000
EMBEDDING_DIM = 16
MAX_LEN = 120
N_UNITS = 32

print("\n--- Loading IMDB Built-in Dataset ---")
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=VOCAB_SIZE)
print(f"Loaded {len(X_train)} training sequences and {len(X_test)} test sequences.")
print("Sequences have varying lengths.")

print("\n--- Preprocessing Data (Padding) ---")
X_train = pad_sequences(X_train, maxlen=MAX_LEN, padding='post', truncating='post')
X_test = pad_sequences(X_test, maxlen=MAX_LEN, padding='post', truncating='post')

print(f"Padded training data (X) shape: {X_train.shape}")
print(f"Padded test data (X) shape: {X_test.shape}")
print(f"Training labels (y) shape: {y_train.shape}")


# -----------------------------------------------------------------------------
# 3. CODE: MODEL BUILDING (FUNCTIONAL API)
# -----------------------------------------------------------------------------


print("\n--- Building Training Model ---")
inputs = Input(shape=(MAX_LEN,), name='input_sequence')

emb_layer = Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM, name='embedding')
embedded_seq = emb_layer(inputs)

bilstm_layer = Bidirectional(
    LSTM(N_UNITS, return_sequences=True),
    name='bidirectional_lstm'
)
rnn_outputs = bilstm_layer(embedded_seq)
attention_layer = Attention(name='attention_layer')
context_vector_sequence = attention_layer([rnn_outputs, rnn_outputs, rnn_outputs])
pooling_layer = GlobalAveragePooling1D(name='context_vector_pooling')
context_vector = pooling_layer(context_vector_sequence)
classifier = Dense(1, activation='sigmoid', name='classifier')
final_output = classifier(context_vector)

training_model = Model(
    inputs,
    final_output,
    name='attention_classification_model'
)

training_model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
print("\n--- Training Model Summary ---")
training_model.summary(line_length=120)
print("\n--- Model Architecture Diagram ---")
try:
    plot_model(
        training_model,
        to_file='attention_model_diagram.png',
        show_shapes=True,
        show_layer_names=True,
        show_layer_activations=True,
        dpi=70
    )
    print("Displaying model architecture diagram (saved as 'attention_model_diagram.png'):")
    display(Image(filename='attention_model_diagram.png'))
except ImportError as e:
    print(f"\nModel visualization failed. {e}")
    print("!pip install pydot graphviz")


# -----------------------------------------------------------------------------
# 4. CODE: MODEL TRAINING
# -----------------------------------------------------------------------------

print("\n--- Training the Attention RNN ---")
history = training_model.fit(
    X_train,
    y_train,
    batch_size=64,
    epochs=5,
    validation_data=(X_test, y_test),
    verbose=1
)

print("Training complete.")


# -----------------------------------------------------------------------------
# 5. CODE: MODEL EVALUATION
# -----------------------------------------------------------------------------

print("\n--- Evaluating Model on Test Data ---")
results = training_model.evaluate(X_test, y_test, verbose=1)
print("\n--- Final Test Results ---")
print(f"Test Loss:     {results[0]:.4f}")
print(f"Test Accuracy: {results[1]:.4f} ({(results[1] * 100):.2f}%)")
```

# Neural Turing Machine (NTM) Implementation

```python
import tensorflow as tf
from tensorflow.keras.layers import Layer, RNN, Dense, Input, Embedding, GRUCell
from tensorflow.keras.models import Sequential
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
import numpy as np
from tensorflow.keras.utils import plot_model
from IPython.display import Image, display

print(f"Using TensorFlow Version: {tf.__version__}")

# -----------------------------------------------------------------------------
# 1. ARCHITECTURE DETAILS & VISUALIZATION
# -----------------------------------------------------------------------------
"""
=================================================
Neural Turing Machine (NTM) - Simplified
=================================================
- **Inventors:** DeepMind (Graves et al.)
- **Year:** 2014
- **Key Concept:** An RNN augmented with an external memory bank,
  which it can learn to read from and write to.

A standard RNN (even LSTM/GRU) has its "memory" fused with its
processing unit (the hidden state). This limits its memory
capacity.

An NTM separates these:
1.  **The Controller:** A standard RNN (like an LSTM or GRU)
    that acts as the "CPU."
2.  **The Memory:** A large external matrix (e.g., N x M) that
    acts as the "RAM."
3.  **The Heads:** One or more "read" and "write" heads that
    interact with the memory.

At each timestep, the controller takes the input AND the data
read from memory in the last step. It then:
1.  Processes this combined information.
2.  Generates a "query" to decide which memory slots to
    read from for the *next* timestep.
3.  (In a full NTM) Generates data and locations to *write*
    new information into the memory.

---
Data Flow Visualization (Simplified Read-Only):
---
                      [ Input (x_t) ] ----+
                                          |
     [ Prev. Read Vector (r_t-1) ] ---+    |
                                     v    v
     [ Prev. Controller State ] -> [CONTROLLER] -> [ Output (y_t) ]
                                (e.g., GRU) |
                                            |
     (Loop for next timestep)               |
     <---------------------------+          v
                                 |    [ Read Query ]
                                 |         |
                                 v         v
     [ New Read Vector (r_t) ] <-[ READ  HEAD ]
                                (Attention)
                                     ^
                                     |
                                [ MEMORY BANK (M) ]

*** DISCLAIMER: ***
A full NTM is extremely complex and not a built-in Keras
layer. This script is a **simplified, conceptual implementation**
for educational purposes. It demonstrates the "controller + memory"
structure using a *read-only* memory with a simple attention
(content-addressing) mechanism.
"""

print("Simplified NTM architecture details loaded.")


# -----------------------------------------------------------------------------
# 2. CODE: CUSTOM NTM CELL
# -----------------------------------------------------------------------------

class NTMCell(Layer):
    """
    A simplified, conceptual NTM cell (read-only).

    This cell's state is a tuple:
    1. The controller's state (from a GRUCell).
    2. The read-vector from the previous timestep.
    """
    def __init__(self, controller_units, memory_size, memory_vector_dim, **kwargs):
        super(NTMCell, self).__init__(**kwargs)
        self.controller_units = controller_units
        self.memory_size = memory_size
        self.memory_vector_dim = memory_vector_dim
        self.controller_cell = GRUCell(self.controller_units)
        self.state_size = (self.controller_units, self.memory_vector_dim)
        self.output_size = self.controller_units

    def build(self, input_shape):
        """Create the weights for the cell."""
        self.memory = self.add_weight(
            shape=(self.memory_size, self.memory_vector_dim),
            initializer='glorot_uniform',
            name='memory_bank'
        )
        self.read_query_layer = Dense(self.memory_vector_dim, name='read_query')
        controller_input_dim = input_shape[-1] + self.memory_vector_dim
        self.controller_cell.build(tf.TensorShape((None, controller_input_dim)))

        self.built = True

    def call(self, inputs, states):
        """
        The core logic of the NTM cell at each timestep.

        Args:
            inputs (tensor): The input at the current timestep (x_t).
            states (list): [prev_controller_state, prev_read_vector]

        Returns:
            A tuple (output, new_state):
            - output (tensor): The controller's output (y_t).
            - new_state (list): [new_controller_state, new_read_vector]
        """
        prev_controller_state = states[0]
        prev_read_vector = states[1]
        combined_input = tf.concat([inputs, prev_read_vector], axis=-1)
        controller_output, [new_controller_state] = self.controller_cell(
            combined_input, [prev_controller_state]
        )
        read_query = self.read_query_layer(controller_output)
        read_query_norm = tf.nn.l2_normalize(read_query, axis=-1)
        memory_norm = tf.nn.l2_normalize(self.memory, axis=-1)

        similarity_scores = tf.matmul(read_query_norm, memory_norm, transpose_b=True)
        read_weights = tf.nn.softmax(similarity_scores, axis=-1)

        new_read_vector = tf.matmul(tf.expand_dims(read_weights, axis=1), self.memory)
        new_read_vector = tf.squeeze(new_read_vector, axis=1) # Squeeze back

        output = controller_output
        new_state = [new_controller_state, new_read_vector]

        return output, new_state

    def get_config(self):
        """For model saving and loading."""
        config = super(NTMCell, self).get_config()
        config.update({
            'controller_units': self.controller_units,
            'memory_size': self.memory_size,
            'memory_vector_dim': self.memory_vector_dim
        })
        return config

print("Custom 'NTMCell' class defined successfully.")


# -----------------------------------------------------------------------------
# 3. CODE: MODEL BUILDING & SUMMARY
# -----------------------------------------------------------------------------


VOCAB_SIZE = 10000
EMBEDDING_DIM = 32
MAX_LEN = 120
CONTROLLER_UNITS = 64
MEMORY_SIZE = 128
MEMORY_VECTOR_DIM = 40

print("\n--- Building Model ---")

ntm_cell_instance = NTMCell(
    controller_units=CONTROLLER_UNITS,
    memory_size=MEMORY_SIZE,
    memory_vector_dim=MEMORY_VECTOR_DIM
)

model = Sequential([
    Input(shape=(MAX_LEN,)),

    Embedding(input_dim=VOCAB_SIZE, output_dim=EMBEDDING_DIM),

    RNN(ntm_cell_instance, return_sequences=False),
    Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
print("\n--- Model Summary ---")
model.summary()
print("\n--- Model Architecture Diagram ---")
try:
    plot_model(
        model,
        to_file='ntm_model_diagram.png',
        show_shapes=True,
        show_layer_names=True,
        show_layer_activations=True,
        dpi=70
    )
    print("Displaying model architecture diagram (saved as 'ntm_model_dimagram.png'):")
    display(Image(filename='ntm_model_diagram.png'))
except ImportError as e:
    print(f"\nModel visualization failed. {e}")
    print("!pip install pydot graphviz")


# -----------------------------------------------------------------------------
# 4. CODE: MODEL TRAINING
# -----------------------------------------------------------------------------

print("\n--- Loading IMDB Built-in Dataset ---")

(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=VOCAB_SIZE)

print("\n--- Preprocessing Data (Padding) ---")
X_train = pad_sequences(X_train, maxlen=MAX_LEN, padding='post', truncating='post')
X_test = pad_sequences(X_test, maxlen=MAX_LEN, padding='post', truncating='post')

print(f"Padded training data (X) shape: {X_train.shape}")
print(f"Padded test data (X) shape: {X_test.shape}")

print("\n--- Training the NTM Network ---")
history = model.fit(
    X_train,
    y_train,
    epochs=5,
    batch_size=32,
    validation_data=(X_test, y_test),
    verbose=1
)
print("Training complete.")


# -----------------------------------------------------------------------------
# 5. CODE: MODEL EVALUATION
# -----------------------------------------------------------------------------

print("\n--- Evaluating Model on Test Data ---")

results = model.evaluate(X_test, y_test, verbose=1)

print("\n--- Final Test Results ---")
print(f"Test Loss:     {results[0]:.4f}")
print(f"Test Accuracy: {results[1]:.4f} ({(results[1] * 100):.2f}%)")
```

# Fast Residual + Attention Bidirectional LSTM (Lite Version)

```python
# =============================================================================
# ‚ö° Fast Residual + Attention Bidirectional LSTM (Lite Version)
# =============================================================================
# Author: Anshuman Sinha
# Goal: 80‚Äì87% accuracy on IMDB dataset
# =============================================================================

import tensorflow as tf
from tensorflow.keras.layers import (
    Input, Embedding, Bidirectional, LSTM,
    Dense, Dropout, Layer, Add, LayerNormalization
)
from tensorflow.keras.models import Model
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import plot_model
from IPython.display import Image, display
import numpy as np

print(f"‚úÖ TensorFlow Version: {tf.__version__}")

# =============================================================================
# 1Ô∏è‚É£ ARCHITECTURE OVERVIEW
# =============================================================================
"""
=================================================
Lite Residual + Attention BiLSTM
=================================================
- Combines bidirectional temporal learning with residual stability
- Adds a simple, efficient attention mechanism

Pipeline:
[ Input ] ‚Üí [ Embedding ] ‚Üí [ BiLSTM1 ] ‚Üí [ BiLSTM2 + Residual ]
‚Üí [ Attention ] ‚Üí [ Dense Layers ] ‚Üí [ Output ]
"""

# =============================================================================
# 2Ô∏è‚É£ ATTENTION LAYER (Lightweight)
# =============================================================================
class LiteAttention(Layer):
    def __init__(self):
        super().__init__()
    def build(self, input_shape):
        self.W = self.add_weight(shape=(input_shape[-1], 1),
                                 initializer='glorot_uniform',
                                 trainable=True)
    def call(self, x):
        e = tf.squeeze(tf.matmul(x, self.W), axis=-1)
        a = tf.nn.softmax(e, axis=1)
        a = tf.expand_dims(a, axis=-1)
        return tf.reduce_sum(x * a, axis=1)

print("üß† Attention layer ready.")

# =============================================================================
# 3Ô∏è‚É£ LOAD & PREPROCESS IMDB DATASET
# =============================================================================
VOCAB_SIZE = 8000
EMBEDDING_DIM = 32
MAX_LEN = 150

print("\nüì¶ Loading IMDB dataset...")
(X_train, y_train), (X_test, y_test) = imdb.load_data(num_words=VOCAB_SIZE)
X_train = pad_sequences(X_train, maxlen=MAX_LEN, padding='post')
X_test = pad_sequences(X_test, maxlen=MAX_LEN, padding='post')
print(f"‚úÖ Data ready ‚Üí Train: {X_train.shape}, Test: {X_test.shape}")

# =============================================================================
# 4Ô∏è‚É£ BUILD MODEL (Compact Version)
# =============================================================================
inputs = Input(shape=(MAX_LEN,))
x = Embedding(VOCAB_SIZE, EMBEDDING_DIM)(inputs)
x = Dropout(0.3)(x)

bilstm1 = Bidirectional(LSTM(64, return_sequences=True))(x)
bilstm2 = Bidirectional(LSTM(64, return_sequences=True))(bilstm1)

# Residual Connection
residual = Add()([bilstm1, bilstm2])
residual = LayerNormalization()(residual)

# Attention
context = LiteAttention()(residual)

# Dense Layers
dense = Dense(64, activation='relu')(context)
dense = Dropout(0.3)(dense)
outputs = Dense(1, activation='sigmoid')(dense)

model = Model(inputs, outputs)
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# =============================================================================
# 5Ô∏è‚É£ MODEL SUMMARY & VISUALIZATION
# =============================================================================
print("\n--- Model Summary ---")
model.summary()

try:
    plot_model(model, to_file='fast_bilstm.png', show_shapes=True, dpi=70)
    display(Image(filename='fast_bilstm.png'))
except Exception as e:
    print("Visualization skipped (install graphviz if needed):", e)

# =============================================================================
# 6Ô∏è‚É£ TRAIN MODEL (Fast Mode)
# =============================================================================
print("\nüöÄ Training started! ")

history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=3,
    batch_size=128,
    verbose=1
)

# =============================================================================
# 7Ô∏è‚É£ EVALUATE MODEL
# =============================================================================
print("\nüìä Evaluating model...")
loss, acc = model.evaluate(X_test, y_test, verbose=0)
print(f"‚úÖ Test Accuracy: {acc*100:.2f}%")
print(f"‚úÖ Test Loss: {loss:.4f}")

# =============================================================================
# 8Ô∏è‚É£ PARAMETER INFO
# =============================================================================
total_params = model.count_params()
trainable_params = np.sum([tf.keras.backend.count_params(p) for p in model.trainable_weights])
non_trainable_params = total_params - trainable_params

print("\n--- Parameter Summary ---")
print(f"Total Params:        {total_params:,}")
print(f"Trainable Params:    {trainable_params:,}")
print(f"Non-trainable Params:{non_trainable_params:,}")

print("\n‚úÖ Done! Training completed!")
```
